```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("C:/Users/LENOVO/Desktop/python/abc_company.csv")

# Display the first few rows of the dataset
df.head()
```

```python
# Randomly generate heights between 150 and 180
df['height'] = np.random.randint(150, 181, size=len(df))

# Check the updated dataframe
df.head()
```

```python
# Task 1: Distribution of employees across each team
team_distribution = df['Team'].value_counts()
team_percentage = team_distribution / len(df) * 100
```

```python
# Display the distribution
print(team_distribution)
```

```python
# Display the percentage
print(team_percentage)
```

```python
# Bar Chart for Team Distribution
plt.figure(figsize=(10, 6))
team_distribution.plot(kind='bar', color='skyblue')
plt.title("Employee Distribution by Team")
plt.xlabel("Team")
plt.ylabel("Number of Employees")
plt.tight_layout()
plt.show()
```

```python
# Pie Chart for Team Percentage
plt.figure(figsize=(8, 8))
team_percentage.plot(kind='pie', autopct='%1.1f%%')
plt.title("Percentage of Employees by Team")
plt.show()

```

```python
# Group by 'Position' and count
position_distribution = df['Position'].value_counts()
```

```python
# Display the positions
print(position_distribution)
```

```python
# Bar Chart for Position Distribution
plt.figure(figsize=(10, 6))
position_distribution.plot(kind='bar', color='lightgreen')
plt.title("Employee Count by Position")
plt.xlabel("Position")
plt.ylabel("Number of Employees")
plt.show()
```

```python
# Stacked Bar Chart for Team vs. Position
pivot_table = df.pivot_table(index='Position', columns='Team', aggfunc='size')

pivot_table.plot(kind='bar', stacked=True, figsize=(12, 7), colormap='viridis')
plt.title("Team and Position Stacked Bar Chart")
plt.ylabel("Number of Employees")
plt.xlabel("Position")
plt.legend(title="Team")
plt.tight_layout()
plt.show()

```

```python
# Define age groups
bins = [20, 30, 40, 50, 60]
labels = ['20-30', '31-40', '41-50', '51-60']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

# Group by age group
age_group_distribution = df['Age Group'].value_counts()

# Display the age group distribution
print(age_group_distribution)
```

```python
# Histogram for Age Distribution
plt.figure(figsize=(10, 6))
df['Age'].plot(kind='hist', bins=5, edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# Line Chart for Age Group Distribution
plt.figure(figsize=(10, 6))
age_group_distribution.sort_index().plot(kind='line', marker='o', color='purple')
plt.title("Age Group Trends")
plt.xlabel("Age Group")
plt.ylabel("Number of Employees")
plt.tight_layout()
plt.show()
```

```python
# Task 4: Team with the highest salary expenditure
team_salary = df.groupby('Team')['Salary'].sum().sort_values(ascending=False)

# Task 4: Position with the highest salary expenditure
position_salary = df.groupby('Position')['Salary'].sum().sort_values(ascending=False)
```

```python
# Display the salary by team
print("Salary by Team:\n", team_salary)
```

```python
# Display the salary by position
print("Salary by Position:\n", position_salary)
```

```python
plt.figure(figsize=(10, 6))
team_salary.plot(kind='barh', color='orange')
plt.title("Total Salary Expenditure by Team")
plt.xlabel("Total Salary Expenditure")
plt.ylabel("Team")
plt.tight_layout()
plt.show()
```

```python
# Plotting the salary by position
plt.figure(figsize=(10,6))
position_salary.plot(kind='bar', color='pink')
plt.title('Salary Expenditure by Position')
plt.xlabel('Position')
plt.ylabel('Total Salary Expenditure')
plt.show()
```

```python
correlation = df['Age'].corr(df['Salary'])
print(f"Correlation between Age and Salary: {correlation:.2f}")
```

```python
# Scatter Plot for Age vs Salary
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Salary'], alpha=0.5, color='purple')
plt.title("Age vs. Salary Correlation")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age Group', y='Salary', data=df, color='yellow')
plt.title("Box Plot of Salary by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Salary")
plt.tight_layout()
plt.show()
```
